```python

import pandas as pd

# Load two csv files
df1 = pd.read_csv('/Compare/fyers-bn-1min.csv')
df2 = pd.read_csv('/compare/kite-bn-1min.csv')
```

```python
df2['DateTime'] = df2['DateTime'].str.split('+').str[0]
df2
```

```python
df1['DateTime'] = pd.to_datetime(df1['DateTime'])
df1
```

```python
df2['DateTime'] = pd.to_datetime(df2['DateTime'])
df2
```

```python
from datetime import datetime

# Create a date using year, month, and day
d = datetime(2023, 5, 31)  # This creates a date for August 12, 2023

df1 = df1[df1['DateTime'] <= d]
```

```python
df2 = df2[df2['DateTime'] <= d]
```

```python
df_merged = pd.merge(df1, df2, on=['DateTime'], how='inner')
df_merged
```

```python
df_merged['Open_diff'] = df_merged['Open_x'] - df_merged['Open_y']
df_merged['High_diff'] = df_merged['High_x'] - df_merged['High_y']
df_merged['Low_diff'] = df_merged['Low_x'] - df_merged['Low_y']
df_merged['Close_diff'] = df_merged['Close_x'] - df_merged['Close_y']
```

```python
df_merged
```

```python
# check total rows of df_merged

len(df_merged)
```

```python
# check open greater than zero
df_merged[df_merged['Open_diff']>0].shape[0]
```

```python
# check open High than zero
df_merged[df_merged['High_diff']>0].shape[0]
```

```python
# check low greater than zero
df_merged[df_merged['Low_diff']>0].shape[0]
```

```python
# check close greater than zero
df_merged[df_merged['Close_diff']>0].shape[0]
```

```python
# Calculate mean
mean_open = df_merged['Open_diff'].mean()
mean_high = df_merged['High_diff'].mean()
mean_low = df_merged['Low_diff'].mean()
mean_close = df_merged['Close_diff'].mean()

# Calculate median
median_open = df_merged['Open_diff'].median()
median_high = df_merged['High_diff'].median()
median_low = df_merged['Low_diff'].median()
median_close = df_merged['Close_diff'].median()

# Calculate mode
mode_open = df_merged['Open_diff'].mode()[0] # mode() returns a Series, use [0] to get the first mode
mode_high = df_merged['High_diff'].mode()[0]
mode_low = df_merged['Low_diff'].mode()[0]
mode_close = df_merged['Close_diff'].mode()[0]

# Print results
print("Mean of Open, High, Low, Close differences:", mean_open, mean_high, mean_low, mean_close)
print("Median of Open, High, Low, Close differences:", median_open, median_high, median_low, median_close)
print("Mode of Open, High, Low, Close differences:", mode_open, mode_high, mode_low, mode_close)
```

```python
import matplotlib.pyplot as plt

# Create a grid of 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot 'Open_diff' data on the top-left subplot
axes[0, 0].plot(df_merged.index, df_merged['Open_diff'], label='Open_diff', color='blue')
axes[0, 0].set_title('Open_diff')
axes[0, 0].set_ylabel('Difference')

# Plot 'High_diff' data on the top-right subplot
axes[0, 1].plot(df_merged.index, df_merged['High_diff'], label='High_diff', color='orange')
axes[0, 1].set_title('High_diff')

# Plot 'Low_diff' data on the bottom-left subplot
axes[1, 0].plot(df_merged.index, df_merged['Low_diff'], label='Low_diff', color='green')
axes[1, 0].set_title('Low_diff')
axes[1, 0].set_ylabel('Difference')
axes[1, 0].set_xlabel('DateTime')

# Plot 'Close_diff' data on the bottom-right subplot
axes[1, 1].plot(df_merged.index, df_merged['Close_diff'], label='Close_diff', color='red')
axes[1, 1].set_title('Close_diff')
axes[1, 1].set_xlabel('DateTime')

# Automatically adjust subplot parameters for better layout
plt.tight_layout()

# Display the plots
plt.show()
```
